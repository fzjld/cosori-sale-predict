```python
import pandas as pd
import xlrd
import pylab as plt
import seaborn as sns
import numpy as np
from fbprophet import Prophet
import json
from fbprophet.serialize import model_to_json, model_from_json
from sklearn.metrics import mean_squared_error, r2_score

%matplotlib inline
```

```python
data = pd.read_csv('./data.csv', encoding='GBK')
data.head(2)
```

```python
data.data_date = pd.to_datetime(data.data_date)
ts = pd.pivot_table(data, index=['asin', 'data_date'], values=['ordered_units'], aggfunc=np.mean)
ts = ts.unstack().T
ts = ts.reset_index()
ts.head(2)
```

## 评价指标

```python
def metrics(y_true, y_pred):
    return 1 - (abs(y_true - y_pred) / y_true).mean()
```

## 对B074NYJL9J建模

```python
ts.B07FKN76ZS.mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(ts.B07FKN76ZS)
```

```python
start = '2019-12-02'
end = '2021-07-22'
a = ts[(ts['data_date'] >= start) & (ts['data_date'] <= end)][['data_date', 'B07FKN76ZS']]
```

```python
y_true = ts[(ts['data_date'] >= '2021-07-23') & (ts['data_date'] <= '2021-8-22')][['data_date', 'B07FKN76ZS']]
```

```python
a.columns = ['ds', 'y']
```

```python
model = Prophet(yearly_seasonality=True)
```

```python
model.fit(a)
```

```python
future = model.make_future_dataframe(periods=31, include_history=False)
```

```python
future.shape
```

```python
y_pred = model.predict(future)
```

```python
y_pred['yhat'].shape
```

```python
mean_squared_error(y_true['B07FKN76ZS'], y_pred['yhat'])
```

```python
pd.DataFrame({'true': y_true['B07FKN76ZS'], 'pred': y_pred['yhat']} )
```

```python
r2_score(y_true['B07FKN76ZS'], y_pred['yhat'])
```

## 对B071D7BTTW建模

```python
plt.figure(figsize=(12,5))
plt.plot(ts.B071D7BTTW)
```

```python
start = '2019-12-02'
end = '2021-07-22'
a = ts[(ts['data_date'] >= start) & (ts['data_date'] <= end)][['data_date', 'B071D7BTTW']]
y_true = ts[(ts['data_date'] >= '2021-07-23') & (ts['data_date'] <= '2021-8-22')][['data_date', 'B071D7BTTW']]

a.columns = ['ds', 'y']
model = Prophet(yearly_seasonality=True)
model.fit(a)
```

```python
future = model.make_future_dataframe(periods=31, include_history=True)
y_pred = model.predict(future)
mean_squared_error(y_true['B071D7BTTW'], y_pred['yhat'])
```

```python
plt.plot(y_pred['yhat'].values)
plt.plot(y_true['B071D7BTTW'].values)
```

```python
def metrics(y_true, y_pred):
    return 1 - (abs(y_true - y_pred).sum()) / y_true.sum()
```

```python
metrics(y_true['B071D7BTTW'].values, y_pred['yhat'].values)
```

```python
r2_score(y_true['B071D7BTTW'], y_pred['yhat'])
```

## 测试：B07FKN76ZS, 电水壶

```python
start = '2019-12-02'
end = '2021-07-22'
train = ts[(ts['data_date'] >= start) & (ts['data_date'] <= end)][['data_date', 'B07FKN76ZS']]
test = ts[(ts['data_date'] >= '2021-07-23') & (ts['data_date'] <= '2021-8-22')]['B07FKN76ZS']
train.columns = ['ds', 'y']

plt.figure(figsize=(12,5))
plt.plot(train.ds, train.y)
```

```python
model = Prophet(yearly_seasonality=True)
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=31, include_history=False)
y_pred = model.predict(future)['yhat']
```

```python
metrics(test.values, y_pred.values)
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.figure(figsize=(12,5))
plt.plot(future.ds, test.values, label='真实销量',)
plt.plot(future.ds, y_pred, label='预测销量')
plt.xticks(rotation=90)
plt.legend()
```
