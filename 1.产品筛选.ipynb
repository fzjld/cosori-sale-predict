```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data.csv', encoding='gbk')
```

```python
df = pd.read_excel('./产品信息 (1).xlsx')
```

```python
# 在售品牌
saling_asins = df[(df['品牌'] == 'COSORI') & (
    df['状态'] == '在售')]['平台商品标识'].unique()
```

```python
saling_asins
```

```python
saling_asins.shape
```

```python
data.asin.nunique()
```

```python
data1 = data[data.asin.isin(saling_asins)]
```

```python
data1.asin.nunique()
```

```python
data.shape, data1.shape
```

```python
# 查看每个产品上线的时间
avg_sales = data1.groupby('asin').agg({'ordered_units': np.average, 'data_date': np.count_nonzero}).sort_values(
    'ordered_units', ascending=False)
avg_sales = avg_sales.rename(
    columns={'ordered_units': 'average_sale', 'data_date': 'sale_date'})
```
